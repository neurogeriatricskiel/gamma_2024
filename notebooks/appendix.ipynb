#### Basics of Python Scripting

**Gait Sequence Detection using KielMAT Module**

We will use the [Paraschiv-Ionescu](https://jneuroengrehab.biomedcentral.com/articles/10.1186/s12984-019-0494-z) gait sequence detection algorithm, implemented in the `kielmat.modules.gsd._paraschiv` package, to detect gait sequences from triaxial accelerometer data collected from a sensor worn on the lower back.

The detection process involves:
- Loading accelerometer data and calculating the norm of the x, y, and z acceleration axes.
- Resampling the data to a 40 Hz sampling frequency and applying smoothing filters to reduce noise.
- Extracting gait-related features using the continuous wavelet transform.
- Identifying active periods with the Hilbert envelope, and using an adaptive threshold to detect steps.
- Organizing detected steps into walking bouts of five or more steps, merging intervals with breaks shorter than three seconds.
- Outputting the detected gait sequences in BIDS format.


**Import Necessary libraries**
  - The necessary libraries such as `numpy`, `matplotlib.pyplot`, keep control dataset, and `Paraschiv-Ionescu` gait sequence detection algorithms will be imported.
  - Make sure that you have all the required libraries and modules installed before running this code.
  - You will also need to install the `kielmat` library and its dependencies if you haven't already.


````python
import numpy as np
import matplotlib.pyplot as plt
import os
from pathlib import Path
from kielmat.datasets.keepcontrol import fetch_dataset, load_recording
from kielmat.modules.gsd import ParaschivIonescuGaitSequenceDetection
````

**Data Preparation**
  - Use `load_recording(...)` to call the load_recording function from the keepcontrol module. It is responsible for downloading the dataset from a remote location associated with the `_keepcontrol` dataset path.
  - Participants performed a series of mobility tasks, such as walking, turning, and sit-to-stand, while wearing the IMUs.
  - The publicaly available dataset includes datasets of two subjects with subjects ids `pp001` and `pp002`.
  - Here, we will take a look at the `pelvis` IMU data from the `walkSlow` trial of subject `pp002`, which is walking a 5 meter level-ground distance at half your normal speed.

````python
# Creating the Dataset Path
keep_control_dataset_path = Path(os.getcwd()) / "_keepcontrol"

# Call the fetch_dataset function from the keepcontrol module to download the dataset associated with the `_keepcontrol` path
fetch_dataset(keep_control_dataset_path)

# The 'load_recording' function is used to load the data from the specified keep_control_dataset_path path
subject_id = "pp002"
task_name = "walkSlow"
recording = load_recording(dataset_path=keep_control_dataset_path, id=subject_id, task=task_name)
````

````python
# Select tracked_point as pelvis since GSD algortihm uses data recorded from lowerback
tracking_sys = 'imu'
tracked_point = "pelvis"

# Load imu data
imu_data = recording.data["imu"]

# Extract lowerback acceleration data
accel_data = imu_data[[f"{tracked_point}_ACCEL_{x}" for x in "xyz"]]

# Get the acceleration unit from the channels
accel_unit = recording.channels[tracking_sys][
    recording.channels[tracking_sys]["name"].str.contains("ACCEL")
]["units"].iloc[0]

# Print acceleration data
print(f"accel_data ({accel_unit}): {accel_data}")
````

````python
# Get the corresponding sampling frequency directly from the recording
fs = recording.channels[tracking_sys]["sampling_frequency"].iloc[0].astype(float)

# Print sampling frequency
print(f"Acceleration data was sampled at: {fs}Hz")
````

**Visualisation of the Data**

The raw acceleration data including components of x, y and z axis is represented.

````python
import matplotlib.pyplot as plt

# Calculate time in minutes
time_in_minutes = [sample / (fs * 60) for sample in range(len(accel_data))]

# Initialize the plot
plt.figure(figsize=(12, 6))

# Loop through each axis (x, y, z) and plot the corresponding data
for axis in "xyz":
    plt.plot(time_in_minutes, accel_data[f"{tracked_point}_ACCEL_{axis}"], label=f'ACCEL_{axis.lower()}')

# Add titles, labels, and legend
plt.title("Lower Back Acceleration Data (Pelvis)")
plt.xlabel("Time (minutes)")
plt.ylabel(f"Acceleration ({accel_unit})")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
````

**Applying Paraschiv-Ionescu Gait Sequence Detection from KielMAT**

Now, we are running Paraschiv-Ionescu gait sequence detection algorithm from gsd module [`KielMAT.kielmat.modules.gsd._paraschiv.ParaschivIonescuGaitSequenceDetection`](https://github.com/neurogeriatricskiel/KielMAT/tree/main/kielmat/modules/gsd/_paraschiv.py) to detect gait sequences.

In order to apply gait sequence detection algorithm, an instance of the ParaschivIonescuGaitSequenceDetection class is created using the constructor, `ParaschivIonescuGaitSequenceDetection()`. The `gsd` variable holds the instance, allowing us to access its methods. The inputs of the algorithm are as follows:

- **Input Data:** `data` consist of accelerometer data (N, 3) for the x, y, and z axes in pandas Dataframe format.
- **Sampling Frequency:** `sampling_freq_Hz` is the sampling frequency of the data.
- **Plot Results:** `plot_results`, if set to True, generates a plot showing the detected gait sequences on the data. The default is False. The onset is represented with the vertical green line and the grey area represents the duration of gait sequence detected by the algorithm.

````python
# Create an instance of the ParaschivIonescuGaitSequenceDetection class
gsd = ParaschivIonescuGaitSequenceDetection()

# Call the gait sequence detection using gsd.detect
gsd = gsd.detect(
    data=accel_data, sampling_freq_Hz=fs, plot_results=True, dt_data=None
)

# Gait sequences are stored in gait_sequences_ attribute of gsd
gait_sequences = gsd.gait_sequences_

# Add events to the recording as a dictionary including tracking system and events
gait_sequence_events = gait_sequences
recording.add_events(tracking_system='imu', new_events=gait_sequence_events)

# Show events and their corresponding information
print(recording.events)
````

**Initial Contact Detection using KielMAT Module**

Now, we are running Paraschiv-Ionescu initial contact detection algorithm from icd module [`KielMAT.kielmat.modules.icd._paraschiv.ParaschivIonescuInitialContactDetection`](https://github.com/neurogeriatricskiel/KielMAT/tree/main/kielmat/modules/icd/_paraschiv.py) to detect initial contacts throughout the detected gait sequences. For this purpose, we have to first apply Paraschiv-Ionescu gait sequences detection algorithm to identify gait sequences using acceleration data.

Then, in order to apply Paraschiv-Ionescu initial contact detection algorithm, an instance of the ParaschivIonescuInitialContactDetection class is created using the constructor, `ParaschivIonescuInitialContactDetection()`. The `icd` variable holds the instance, allowing us to access its methods. The inputs of Paraschiv-Ionescu initial contact detection algorithm are as follows:

- **Input Data:** `data` consist of accelerometer data (N, 3) for the x, y, and z axes in pandas Dataframe format.
- **Gait Sequences:** `gait_sequences`, consist of gait sequences detected by Paraschiv gait sequence detection ([`KielMAT.kielmat.modules.gsd._paraschiv.ParaschivIonescuGaitSequenceDetection`](https://github.com/neurogeriatricskiel/KielMAT/tree/main/kielmat/modules/gsd/_paraschiv.py)).
- **Sampling Frequency:** `sampling_freq_Hz` is the sampling frequency of the data, defined in Hz, with a default value of 100 Hz.
- **Name of vertical acceleration:** `v_acc_col_name` is the name of the vertical acceleration.



````python
from kielmat.modules.icd import ParaschivIonescuInitialContactDetection

# Now, use Paraschiv-Ionescu initial contact detection algortihm to find initial contacts within detected gait sequences.
icd = ParaschivIonescuInitialContactDetection()

# Call the initial contact detection using icd.detect
icd = icd.detect(
    data=accel_data,
    gait_sequences=gait_sequences,
    sampling_freq_Hz=fs,
    v_acc_col_name="pelvis_ACCEL_x"
)

# Print initial contacts information
print(icd.initial_contacts_)
````

**Visualization of the Detected Initial Contacts**

In the following, the raw data of the lower back sensor is plotted with the detected events. The events are plotted as vertical lines. The events are:
- **Gait onset**: Start of the gait sequence
- **Gait duration**: Duration of the gait sequence
- **Initial contacts**: Initial contacts

The gait onset is represented with the vertical green line and the grey area represents the duration of gait sequence detected by the algorithm. The vertical dashed blue lines are representing detected initial contacts within first gait sequence.

````python
# Access the first detected gait sequence
gait_sequence = gsd.gait_sequences_[gsd.gait_sequences_["event_type"] == "gait sequence"].iloc[0]

# Print information about initial contacts within the first gait sequence
ic_within_gait = icd.initial_contacts_[
    icd.initial_contacts_["onset"].between(
        gait_sequence["onset"],
        gait_sequence["onset"] + gait_sequence["duration"],
    )
]
print("\nInitial Contacts within the First Gait Sequence:", ic_within_gait)

# Plot the raw data from the lower back
fig, ax = plt.subplots(figsize=(16, 10))

# Plot raw acceleration data
for axis in "xyz":
    plt.plot(time_in_minutes, accel_data[f"{tracked_point}_ACCEL_{axis}"], label=f'ACCEL_{axis.lower()}')

# Plot the first element of gait sequences
plt.axvline((gait_sequence["onset"]/60), color="g", label="Gait onset")
ax.axvspan(
    gait_sequence["onset"]/60,
    (gait_sequence["onset"] + gait_sequence["duration"])/60,
    alpha=0.2,
    color="gray",
    label="Gait duration",
)

# Plot the initial contacts within the first gait sequence
for ic_time in (ic_within_gait["onset"]/60):
    ax.axvline(ic_time, color="blue", linestyle="--")

# Customize plot
ax.set_xlabel("Time (min)", fontsize=20)
ax.set_ylabel("Acceleration (g)", fontsize=20)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax.legend(
    ["ACCEL x", "ACCEL y", "ACCEL z", "Gait onset", "Gait duration", "Initial contacts"],
    fontsize=20,
    loc="upper right",
)
plt.show()
````
