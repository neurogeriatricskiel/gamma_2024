**Example**

Apply a Butterworth low-pass filter to the artifical data with specified cutoff frequency, sampling rate and filter order. Then, plot the the data and filtered data together.

- Import libraries:
   - Import `numpy` library as `np` for generating artificial data.
   - Import Butterworth filter with `butter` from `scipy.signal` library. This is used to design an Nth-order digital or analog Butterworth filter and return the filter coefficients.
   - Import filtfilt filter with `filtfilt` from `scipy.signal` library. This will apply a digital filter forward and backward to a signal.
   - Import `matplotlib.pyplot` as `plt` for plotting the original and filtered data.

`````python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
`````

- Define a function to filter data using 4th order Butterworth filter with 20Hz cut off frequency and 100Hz sampling rate:
   - use `butter(N, Wn, btype='low', analog=False, output='ba', fs=None)`. Where `N` is oerder of filter, `Wn` is the critical frequency in Hz, `btype` is the type of filter and when `analog` is `False`, return an digital filter, otherwise an analog filter is returned.
   - Note that you need to normalize the cutoff frequency by the Nyquist frequency which gives a value between 0 and 1 to give you the critical frequency `Wn` which is required by the Butterworth filter.
   - Also add docstrings to the function using """        """.

`````python
# Define a function to apply a Butterworth low-pass filter to the input data with specified cutoff frequency, sampling rate and filter order
def butter_lowpass_filter(data, cutoff, fs, order=4):
    """
    Applies a Butterworth low-pass filter to the input data.

    Parameters:
    - data: array-like
        The input signal data to be filtered.
    - cutoff: float
        The cutoff frequency of the filter (in Hz).
    - fs: float
        The sampling frequency of the input data (in Hz).
    - order: int, optional
        The order of the Butterworth filter. Higher order means a sharper cutoff.
        Default is 4.

    Returns:
    - filtered_data: array-like
        The filtered signal.
    """

    # Calculate the Nyquist frequency, which is half the sampling frequency
    nyq = 0.5 * fs

    # Normalize the cutoff frequency by the Nyquist frequency
    # This gives a value between 0 and 1, which is required by the Butterworth filter
    normal_cutoff = cutoff / nyq

    # Design a Butterworth low-pass filter with the specified order and normalized cutoff frequency
    b, a = butter(N=order, Wn=normal_cutoff, btype='low', analog=False)

    # Apply the filter to the input data using filtfilt, which performs forward and backward filtering
    # This helps to minimize phase distortion
    filtered_data = filtfilt(b, a, data)

    return filtered_data

# Assuming 'data' is your time series data, then you can apply this Butterworth low-pass filter
#filtered_data = butter_lowpass_filter(data, cutoff=6, fs=100, order=4)
`````

- Create artifical data to filter with the lowpass filter created above
  - Data should be 1000 samples and should resemble joint angles
  - Create `time` array using `np.linspace` which generates 1000 equally spaced values between 0 and 10 seconds.
  - This represents the time over which our signal (joint angle data) will be sampled.

`````python
time = np.linspace(0, 10, 1000)
`````

- Generate artificial data representing joint angles
  - The data is a combination of two sine waves with different frequencies (2 Hz and 10 Hz).
  - Added random noise with `np.random.normal` to simulate measurement noise. This gives us a noisy signal to filter.


`````python
data = 10 * np.sin(2 * np.pi * 2 * time) + 5 * np.sin(2 * np.pi * 10 * time) + np.random.normal(0, 1, 1000)
`````

- Apply the low-pass filter to the created artifical data
  - Use the butter_lowpass_filter function defined earlier to filter the data.
  - Set cutoff frequency to 6 Hz, sampling frequency (fs) to 100 Hz, and filter order to 4.
  - The filtered_data variable now contains the smoothed version of the joint angle data.

`````python
filtered_data = butter_lowpass_filter(data, 6, 100, order=4)
`````

- Plot the original and filtered data
  - Create a plot with a desired size (i.e. 10 by 6 inches).
  - Plot the original noisy data in one color and the filtered data in another color.
  - Add labels, a title, and a legend for clarity.
  - Display the grid to make it easier to analyze the data visually.


`````python
# Set up the figure for the plot
# Define the figure size as 10 by 6 inches for better readability.
plt.figure(figsize=(10, 6))

# Plot the original noisy data
# - Plot time on the x-axis and the original data on the y-axis.
# - Label it as 'Original Data' for the legend to distinguish it from the filtered data.
plt.plot(time, data, color='red', label='Original Data')

# Plot the filtered data on the same figure
# - Add another plot with the filtered data and label it as 'Filtered Data'.
plt.plot(time, filtered_data, color='blue', label='Filtered Data')

# Add axis labels and title to provide context for the plot
# - x-axis labeled 'Time (s)' indicating time in seconds.
# - y-axis labeled 'Joint Angle (degrees)' indicating the signal type.
# - Title describes the plot content: Original and Filtered Joint Angle Data.
plt.xlabel('Time (s)')
plt.ylabel('Joint Angle (degrees)')
plt.title('Original and Filtered Joint Angle Data')

# Add a legend to identify the original and filtered data lines
# - This helps distinguish between the two data sets on the plot.
plt.legend()

# Display a grid to make it easier to analyze the data visually
# - The grid adds horizontal and vertical lines for reference.
plt.grid(True)

# Display the plot
# - Finally, show the plot with all elements included.
plt.show()
`````

**Gait Sequence Detection using KielMAT**

We will use the [Paraschiv-Ionescu](https://jneuroengrehab.biomedcentral.com/articles/10.1186/s12984-019-0494-z) gait sequence detection algorithm, implemented in the `kielmat.modules.gsd._paraschiv` package, to detect gait sequences from triaxial accelerometer data collected from a sensor worn on the lower back.

The detection process involves:
- Loading accelerometer data and calculating the norm of the x, y, and z acceleration axes.
- Resampling the data to a 40 Hz sampling frequency and applying smoothing filters to reduce noise.
- Extracting gait-related features using the continuous wavelet transform.
- Identifying active periods with the Hilbert envelope, and using an adaptive threshold to detect steps.
- Organizing detected steps into walking bouts of five or more steps, merging intervals with breaks shorter than three seconds.
- Outputting the detected gait sequences in BIDS format.

**Applying Paraschiv-Ionescu Gait Sequence Detection from KielMAT**

Now, we are running Paraschiv-Ionescu gait sequence detection algorithm from gsd module [`KielMAT.kielmat.modules.gsd._paraschiv.ParaschivIonescuGaitSequenceDetection`](https://github.com/neurogeriatricskiel/KielMAT/tree/main/kielmat/modules/gsd/_paraschiv.py) to detect gait sequences.

- **Plot Results:** `plot_results`, if set to True, generates a plot showing the detected gait sequences on the data. The default is False. The onset is represented with the vertical green line and the grey area represents the duration of gait sequence detected by the algorithm.

**Import Necessary libraries**
  - Import necessary libraries such as `numpy`, `pandas` and `matplotlib.pyplot`.

  - Import GaitMap Utilities:
    - **`rotate_dataset`**: 
      - A function for rotating motion data based on specified angles and axes.
      - Important for aligning data captured from different sensors.
    - **`rotation_from_angle`**: 
      - Generates a rotation matrix given an angle and axis.
      - Helps in the transformation of coordinate systems.

  - Import GaitMap Coordinate Conversion:
    - **`convert_to_fbf`**: 
      - Converts the dataset to a foot-based frame (FBF) for standardized gait analysis.
      - Facilitates the interpretation of data relative to the foot's position.

  - Import Stride Segmentation:
    - **`BarthDtw`**: 
      - A class for segmenting strides using Dynamic Time Warping (DTW).
      - Essential for identifying and separating different gait cycles within the data.

  - Import KielMAT Datasets:
    - **`keepcontrol`**: 
      - A module for loading and managing pre-existing datasets.
      - Provides access to motion capture datasets suitable for analysis.

`````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import os
from gaitmap.utils.rotations import rotate_dataset, rotation_from_angle
from gaitmap.utils.coordinate_conversion import convert_to_fbf
from gaitmap.stride_segmentation import BarthDtw
from kielmat.datasets import keepcontrol
`````

**Load Dataset**
  - We use the keepcontrol library to fetch and load the keep control dataset.
  - The data contains information from two sensors - one from the right and one from the left foot.

`````python
# Creating the Dataset Path
keep_control_dataset_path = Path(os.getcwd()) / "_keepcontrol"

# Define tracking system and tracked points
tracking_sys = "imu"
tracked_points = {tracking_sys: ["left_foot", "right_foot"]}

# Load recording data
recording = keepcontrol.load_recording(
    id="pp002",
    task="walkSlow",
    tracking_systems=[tracking_sys],
    tracked_points=tracked_points
)

# Extract IMU sampling rate from recording
fs = recording.channels["imu"]["sampling_frequency"].unique()[0]
`````

**Preprocess the Data**

In this section, we preprocess the IMU data for the left and right foot sensors by performing the following steps:

- **Extract IMU Data**: We extract the IMU data from the recording for both left and right foot sensors.
  
- **Define Rotation**: A rotation is defined for the right foot sensor, specifically -180 degrees around the x-axis, to align its orientation with the left foot's data.

- **Create MultiIndex DataFrame**: 
   - We prepare a MultiIndex DataFrame by splitting each column name into two components: 'sensor' (identifying left or right) and 'axis' (specifying the measurement axis like gyroscope or accelerometer).
   - This MultiIndex structure allows for hierarchical organization of the data, making it easier to manipulate and analyze.

- **Assign MultiIndex**: We assign this MultiIndex to the IMU data DataFrame columns.

- **Display Data**: Finally, we display the first few rows of the formatted IMU data to verify our preprocessing steps.


`````python
# Extract IMU data for the left and right foot sensors
imu_data = recording.data[tracking_sys]

# Define the rotation for the right foot sensor
right_sensor_rotation = rotation_from_angle(
    np.array([0, 0, 1]), np.deg2rad(-180)  # -180 degrees around the x-axis
)

# Prepare to create a MultiIndex DataFrame for easier data manipulation
# Split each column name into 'sensor' and 'axis' components
split_column_names = [
    (col.split('_', 2)[0] + '_sensor', col.split('_', 2)[2][:3].lower() + '_' + col.split('_', 2)[2][-1])
    for col in imu_data.columns
]

# Create a MultiIndex using the split column names
multiindex_columns = pd.MultiIndex.from_tuples(split_column_names, names=["sensor", "axis"])

# Assign the MultiIndex to the IMU data DataFrame columns
imu_data.columns = multiindex_columns

# Display the first few rows of the formatted IMU data
print(imu_data.head())
`````

**Filter and Transform IMU Data**

In this section, we further process the IMU data to prepare it for analysis by performing the following steps:

- **Remove Magnetic Field Data**: 
   - We filter out any columns from the IMU data that contain magnetic field measurements (identified by the "mag" in their axis names). This step helps us focus only on relevant data for our analysis.

- **Rotate Right Foot Sensor Data**: 
   - We apply the previously defined rotation matrix to the right foot sensor data to ensure that all sensor data is aligned correctly in the same frame of reference.

- **Convert to Foot-Frame**: 
   - The rotated data is then converted to the foot-frame of reference, allowing us to analyze the movements of each foot in a consistent manner.

- **Display Processed Data**: 
   - Finally, we display the first few rows of the processed foot-frame data to verify the transformations applied.

`````python
# Remove all columns containing magnetic field ("mag") data
imu_data_filtered = imu_data.loc[:, ~imu_data.columns.get_level_values("axis").str.contains("mag")]

# Rotate the right foot sensor data using the defined rotation matrix
rotations = dict(right_sensor=right_sensor_rotation)
imu_data_rotated = rotate_dataset(imu_data_filtered, rotations)

# Convert the rotated data to the foot-frame of reference
foot_frame_data = convert_to_fbf(imu_data_rotated, left_like="left_", right_like="right_")

# Display the first few rows of the processed data
print(foot_frame_data.head())
`````

**Results of IMU Data Processing**

The table displays the processed IMU data for both the left and right foot sensors after applying the necessary transformations.

- **Acceleration Data**:
  - `acc_pa`: Acceleration in the anterior-posterior direction.
  - `acc_ml`: Acceleration in the medio-lateral direction.
  - `acc_si`: Acceleration in the superior-inferior direction.

- **Gyroscope Data**:
  - `gyr_pa`: Angular velocity in the anterior-posterior direction.
  - `gyr_ml`: Angular velocity in the medio-lateral direction.
  - `gyr_si`: Angular velocity in the superior-inferior direction.

**Applying the DTW**

In this section, we utilize Dynamic Time Warping (DTW) to analyze our IMU data and identify similarities between time-series sequences.

- **Initialization**: We start by initializing the Barth Dynamic Time Warping (DTW) algorithm. This step sets up the DTW framework for analyzing the data:

- **Segmenting the Data**: Once the DTW algorithm is initialized, we apply it to segment the foot-frame data using the IMU sampling rate. This process involves aligning the sequences and identifying temporal patterns in the data:

`````python
# Initialize the Barth Dynamic Time Warping (DTW) algorithm
dtw = BarthDtw()

# Apply DTW for segmenting the data using the IMU sampling rate
dtw = dtw.segment(data=foot_frame_data, sampling_rate_hz=fs)
`````

**Inspecting the Results**

After applying Dynamic Time Warping (DTW) to our IMU data, we can inspect the results to understand the detected strides better.

- The primary output of the DTW process is the `stride_list_`, which contains the start and end times of all identified strides. 
- Since we provided a dataset with two sensors, the output will be in the form of a dictionary.

`````python
# Extract the stride list for the left foot sensor
stride_list_left = dtw.stride_list_["left_sensor"]

# Print the number of detected strides
print(f"{len(stride_list_left)} strides were detected.")

# Display the first few rows of the detected stride list
stride_list_left.head()
`````

**Visualizing DTW Results**

To gain a deeper understanding of the results from our Dynamic Time Warping (DTW) analysis, we can plot additional information regarding the detected strides and the DTW cost function.

- Plot the gyroscope data of the left foot sensor, the DTW cost function, and highlight the identified strides.

`````python
# Define the sensor to be used for plotting (left foot sensor)
sensor = "left_sensor"

# Create subplots for gyroscope data and DTW cost function
fig, axs = plt.subplots(nrows=2, sharex=True, figsize=(12, 6))

# Plot the gyroscope data for the medio-lateral axis
dtw.data[sensor]["gyr_ml"].reset_index(drop=True).plot(ax=axs[0], color="blue")
axs[0].set_ylabel("Gyro [deg/s]")

# Plot the DTW cost function
axs[1].plot(dtw.cost_function_[sensor], color="purple")
axs[1].set_ylabel("DTW Cost [a.u.]")
axs[1].legend()

# Add stride annotations (green shaded areas) to the gyroscope and cost function plots
for _, s in dtw.stride_list_[sensor][["start", "end"]].iterrows():
    axs[0].axvspan(*s, alpha=0.3, color="green", label="Stride")  # Highlight strides in gyroscope plot
    axs[1].axvspan(*s, alpha=0.3, color="green")  # Highlight strides in cost function plot

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
`````
