**Example**

Apply a Butterworth low-pass filter to the artifical data with specified cutoff frequency, sampling rate and filter order. Then, plot the the data and filtered data together.

- Import libraries:
   - Import `numpy` library as `np` for generating artificial data.
   - Import Butterworth filter with `butter` from `scipy.signal` library. This is used to design an Nth-order digital or analog Butterworth filter and return the filter coefficients.
   - Import filtfilt filter with `filtfilt` from `scipy.signal` library. This will apply a digital filter forward and backward to a signal.
   - Import `matplotlib.pyplot` as `plt` for plotting the original and filtered data.

`````python
import numpy as np
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt
`````

- Define a function to filter data using 4th order Butterworth filter with 20Hz cut off frequency and 100Hz sampling rate:
   - use `butter(N, Wn, btype='low', analog=False, output='ba', fs=None)`. Where `N` is oerder of filter, `Wn` is the critical frequency in Hz, `btype` is the type of filter and when `analog` is `False`, return an digital filter, otherwise an analog filter is returned.
   - Note that you need to normalize the cutoff frequency by the Nyquist frequency which gives a value between 0 and 1 to give you the critical frequency `Wn` which is required by the Butterworth filter.
   - Also add docstrings to the function using """        """.

`````python
# Define a function to apply a Butterworth low-pass filter to the input data with specified cutoff frequency, sampling rate and filter order
def butter_lowpass_filter(data, cutoff, fs, order=4):
    """
    Applies a Butterworth low-pass filter to the input data.

    Parameters:
    - data: array-like
        The input signal data to be filtered.
    - cutoff: float
        The cutoff frequency of the filter (in Hz).
    - fs: float
        The sampling frequency of the input data (in Hz).
    - order: int, optional
        The order of the Butterworth filter. Higher order means a sharper cutoff.
        Default is 4.

    Returns:
    - filtered_data: array-like
        The filtered signal.
    """

    # Calculate the Nyquist frequency, which is half the sampling frequency
    nyq = 0.5 * fs

    # Normalize the cutoff frequency by the Nyquist frequency
    # This gives a value between 0 and 1, which is required by the Butterworth filter
    normal_cutoff = cutoff / nyq

    # Design a Butterworth low-pass filter with the specified order and normalized cutoff frequency
    b, a = butter(N=order, Wn=normal_cutoff, btype='low', analog=False)

    # Apply the filter to the input data using filtfilt, which performs forward and backward filtering
    # This helps to minimize phase distortion
    filtered_data = filtfilt(b, a, data)

    return filtered_data

# Assuming 'data' is your time series data, then you can apply this Butterworth low-pass filter
#filtered_data = butter_lowpass_filter(data, cutoff=6, fs=100, order=4)
`````

- Create artifical data to filter with the lowpass filter created above
  - Data should be 1000 samples and should resemble joint angles
  - Create `time` array using `np.linspace` which generates 1000 equally spaced values between 0 and 10 seconds.
  - This represents the time over which our signal (joint angle data) will be sampled.

`````python
time = np.linspace(0, 10, 1000)
`````

- Generate artificial data representing joint angles
  - The data is a combination of two sine waves with different frequencies (2 Hz and 10 Hz).
  - Added random noise with `np.random.normal` to simulate measurement noise. This gives us a noisy signal to filter.


`````python
data = 10 * np.sin(2 * np.pi * 2 * time) + 5 * np.sin(2 * np.pi * 10 * time) + np.random.normal(0, 1, 1000)
`````

- Apply the low-pass filter to the created artifical data
  - Use the butter_lowpass_filter function defined earlier to filter the data.
  - Set cutoff frequency to 6 Hz, sampling frequency (fs) to 100 Hz, and filter order to 4.
  - The filtered_data variable now contains the smoothed version of the joint angle data.

`````python
filtered_data = butter_lowpass_filter(data, 6, 100, order=4)
`````

- Plot the original and filtered data
  - Create a plot with a desired size (i.e. 10 by 6 inches).
  - Plot the original noisy data in one color and the filtered data in another color.
  - Add labels, a title, and a legend for clarity.
  - Display the grid to make it easier to analyze the data visually.


`````python
# Set up the figure for the plot
# Define the figure size as 10 by 6 inches for better readability.
plt.figure(figsize=(10, 6))

# Plot the original noisy data
# - Plot time on the x-axis and the original data on the y-axis.
# - Label it as 'Original Data' for the legend to distinguish it from the filtered data.
plt.plot(time, data, color='red', label='Original Data')

# Plot the filtered data on the same figure
# - Add another plot with the filtered data and label it as 'Filtered Data'.
plt.plot(time, filtered_data, color='blue', label='Filtered Data')

# Add axis labels and title to provide context for the plot
# - x-axis labeled 'Time (s)' indicating time in seconds.
# - y-axis labeled 'Joint Angle (degrees)' indicating the signal type.
# - Title describes the plot content: Original and Filtered Joint Angle Data.
plt.xlabel('Time (s)')
plt.ylabel('Joint Angle (degrees)')
plt.title('Original and Filtered Joint Angle Data')

# Add a legend to identify the original and filtered data lines
# - This helps distinguish between the two data sets on the plot.
plt.legend()

# Display a grid to make it easier to analyze the data visually
# - The grid adds horizontal and vertical lines for reference.
plt.grid(True)

# Display the plot
# - Finally, show the plot with all elements included.
plt.show()
`````

`````python
import requests  # Library for making HTTP requests to fetch data from URLs
import c3d      # Library for reading and processing C3D files used in motion capture
import numpy as np  # Fundamental package for numerical operations and handling arrays
import matplotlib.pyplot as plt  # Library for creating visualizations and plots
from io import BytesIO  # Module for handling binary data as in-memory byte streams

# URL of the raw .c3d file on GitHub
url = "https://github.com/neurogeriatricskiel/gamma_2024/raw/main/datasets/c3d%20data/sample_data.c3d"

# Fetch the file from GitHub
response = requests.get(url)  # Make an HTTP GET request to retrieve the C3D file
`````

- Store Data
  - The file content is accessed using a binary reader, which the `c3d` library can read and process.
  - Next, we will convert the point data from each frame into a NumPy array for for further analysis.

`````python
# Initialize an empty list to store all the point data from the C3D file
all_points_data = []

# Open the C3D file in binary read mode using BytesIO
with BytesIO(response.content) as handle:
    # Create a reader object for reading frames from the C3D file
    reader = c3d.Reader(handle)

    # Iterate through each frame in the C3D file
    for frame_no, points, _ in reader.read_frames():
        # Append the point data (3D coordinates of markers) from each frame to the list
        all_points_data.append(points)

# Convert the list of point data into a NumPy array for easier manipulation and analysis
all_points_array = np.array(all_points_data)

# Print the shape of the data structure
print("Shape of all_points_array:", all_points_array.shape)
`````

  - Summary of Output
    - 142 frames of data are available.
    - Each frame has data for 77 markers.
    - Each marker's data consists of 5 attributes (typically three positional coordinates and two additional attributes).

`````python
# Find the index of 'A22:RKneeAngles' in the point_labels array
# np.where returns the indices where the condition is true (where the label matches 'A22:RKneeAngles')
r_knee_angles_index = np.where(point_labels == 'A22:RKneeAngles               ')[0]

# Find the index of 'A22:LKneeAngles' in the point_labels array
# This does the same thing as above, but for the left knee angles
l_knee_angles_index = np.where(point_labels == 'A22:LKneeAngles               ')[0]

# Get the first index of the right knee data
# Since np.where() returns an array, this ensures we only take the first match (index 0)
r_knee_angles_index = r_knee_angles_index[0]

# Get the first index of the left knee data
# Similarly, this retrieves the first match for the left knee data
l_knee_angles_index = l_knee_angles_index[0]

# Extract the x-axis (first dimension, i.e., angle data along the x-direction) 
# for the 'RKneeAngles' from the point data array
# all_points_array[:, r_knee_angles_index, 0] selects the first dimension (x-axis) of the data 
# across all frames for the right knee angles
x_data_r_knee_angles = all_points_array[:, r_knee_angles_index, 0]

# Similarly, extract the x-axis (first dimension) for the 'LKneeAngles' from the point data array
x_data_l_knee_angles = all_points_array[:, l_knee_angles_index, 0]

# Plot the extracted right and left knee angles over time
plt.figure(figsize=(10, 5))
plt.plot(x_data_r_knee_angles, label='Right Knee Angles', color='blue')
plt.plot(x_data_l_knee_angles, label='Left Knee Angles', color='red')
plt.xlabel('Frame')
plt.ylabel('Knee angle')
plt.grid()
plt.legend()
plt.show()
`````

Breakdown of the Output

1. **`name`**:
   - This column indicates the name of the measurement channel. For example, `head_ACCEL_x` refers to the acceleration measurement along the x-axis for the head.

2. **`component`**:
   - This column specifies the component of the measurement. In the case of acceleration channels, the components are typically `x`, `y`, and `z` (representing three-dimensional space).

3. **`type`**:
   - This column indicates the type of sensor measurement. For instance, `ACCEL` denotes that the channel corresponds to acceleration data, while `GYRO` indicates that it corresponds to gyroscope data.

4. **`tracked_point`**:
   - This column specifies the point being tracked by the sensor. In this case, all entries are labeled as `head`, indicating that the data is being collected for the head of the subject.

5. **`units`**:
   - This column describes the units of measurement for the data. For acceleration, the unit is typically in grams (`g`), while for gyroscope data, it is in degrees per second (`deg/s`).

6. **`sampling_frequency`**:
   - This column indicates the rate at which data is sampled. Here, a sampling frequency of `200` means that the measurements are taken 200 times per second.

7. **`tracking_system`**:
   - This column identifies the tracking system used. In this case, it is labeled as `imu`, which stands for Inertial Measurement Unit.

- Create a mapping of each channel type to its associated units using a dictionary comprehension
  -  This information helps you understand the physical quantities being measured and their units, which are important for accurate data interpretation and processing.

`````python
# Create a dictionary mapping each channel type to its units using a dictionary comprehension.
mapping_type_units = {
    ch_type: imu_channels[imu_channels["type"] == ch_type]["units"].iloc[0]
    for ch_type in imu_channels["type"].unique()
}

# Print the channel type-to-units mapping.
print(f"The mapping of channel types to units is: {mapping_type_units}")
`````

- Explanation of the code:
  - `imu_channels["type"].unique()` returns all unique types of channels (e.g., ACCEL, GYRO, MAGN).
  - The dictionary comprehension iterates over each unique `ch_type`.
  - For each `ch_type`, it filters the DataFrame `imu_channels` to rows where the "type" is equal to `ch_type`.
  - It then selects the "units" column and retrieves the first unit associated with each type using `.iloc[0]`.
  - The result is a dictionary `mapping_type_units` that maps each channel type to its unit as:
   
    `{'ACCEL': 'g', 'GYRO': 'deg/s', 'MAGN': 'Gauss'}`.

**Estimating Orientation of the IMU in a Global Coordinate System**

To estimate the orientation of the IMU in a global coordinate system, we can fuse data from the accelerometer and gyroscope using a sensor fusion algorithm. Here, we will use the versatile quaternion filter (VQF, Laidig and Seel, 2023).

Let us take a quick look at the documentation: [VQF Documentation](https://vqf.readthedocs.io/en/latest/getting_started.html).

Thes following steps are crucial for preparing the data before applying VQF orientation estimation.

  - **Data Extraction**: Retrieve the relevant acceleration and gyro data for the specified tracked points (left and right feet) and format them as NumPy arrays for efficient numerical computations.

`````python
# Specify the tracked points
left_foot = "left_foot"
right_foot = "right_foot"

# Extract gyroscope and accelerometer data for the left and right foot
gyr_left = imu_data[[f"{left_foot}_GYRO_{x}" for x in "xyz"]].to_numpy()
acc_left = imu_data[[f"{left_foot}_ACCEL_{x}" for x in "xyz"]].to_numpy()

gyr_right = imu_data[[f"{right_foot}_GYRO_{x}" for x in "xyz"]].to_numpy()
acc_right = imu_data[[f"{right_foot}_ACCEL_{x}" for x in "xyz"]].to_numpy()
`````

  - **Memory Management**: By ensuring that the data is stored in a contiguous array format, the code optimizes performance for subsequent calculations.

    **Note**: In Python, arrays stored in memory can either be contiguous or non-contiguous. Contiguous memory means that the elements of the array are stored sequentially, without gaps, in a single block of memory. This is important for performance, particularly when performing numerical computations, as contiguous arrays allow for faster processing and access by the underlying hardware or libraries like NumPy.

`````python
# Ensure the data is in contiguous memory
gyr_left = np.ascontiguousarray(gyr_left, dtype=np.float64)
acc_left = np.ascontiguousarray(acc_left, dtype=np.float64)
gyr_right = np.ascontiguousarray(gyr_right, dtype=np.float64)
acc_right = np.ascontiguousarray(acc_right, dtype=np.float64)
`````

  - **Unit Conversion**: We need to performs necessary conversions to ensure that the data is in the correct units for further processing with VQF. Gyroscope data is converted from `deg/s` to `rad/s`, and accelerometer data is converted from `g` to `m/s²`.

`````python
# If necessary, convert gyroscope data from degrees per second to radians per second
gyr_left = gyr_left * np.pi / 180 if mapping_type_units["GYRO"] == "deg/s" else gyr_left
gyr_right = gyr_right * np.pi / 180 if mapping_type_units["GYRO"] == "deg/s" else gyr_right

# If necessary, convert accelerometer data from 'g' to meters per second squared
acc_left = acc_left * 9.81 if mapping_type_units["ACCEL"] == "g" else acc_left
acc_right = acc_right * 9.81 if mapping_type_units["ACCEL"] == "g" else acc_right
`````

- **Orientation estimation using a versatile quaternion filter (VQF)**: The original acceleration and gyroscope data are rotated using these estimated orientations, resulting in new datasets (`acc_rot_left`, `gyr_rot_left`, `acc_rot_right`, `gyr_rot_right`) that represent the sensor readings in a consistent orientation frame.

`````python
# Define the sample time based on the sampling frequency
Ts = 1 / fs

# Initialize the versatile quaternion filter for left and right foot IMU data
vqf_left = vqf.VQF(Ts)
vqf_right = vqf.VQF(Ts)

# Run the orientation estimation for the left foot
out_left = vqf_left.updateBatch(gyr_left, acc_left)

# Run the orientation estimation for the right foot
out_right = vqf_right.updateBatch(gyr_right, acc_right)

# 'out_left' and 'out_right' will now contain the estimated quaternions for both feet. # Initialize arrays to hold the rotated acceleration and gyroscope data for left and right foot
acc_rot_left = np.zeros_like(acc_left)  # Array for rotated acceleration data for left foot
gyr_rot_left = np.zeros_like(gyr_left)  # Array for rotated gyroscope data for left foot

acc_rot_right = np.zeros_like(acc_right)  # Array for rotated acceleration data for right foot
gyr_rot_right = np.zeros_like(gyr_right)  # Array for rotated gyroscope data for right foot

# Loop through each time point to perform quaternion rotation for the left foot
for t in range(len(acc_left)):
    # Rotate the acceleration vector using the estimated quaternion at time t for the left foot
    acc_rot_left[t] = vqf_left.quatRotate(out_left["quat6D"][t], acc_left[t])

    # Rotate the gyroscope vector using the estimated quaternion at time t for the left foot
    gyr_rot_left[t] = vqf_left.quatRotate(out_left["quat6D"][t], gyr_left[t])

# Loop through each time point to perform quaternion rotation for the right foot
for t in range(len(acc_right)):
    # Rotate the acceleration vector using the estimated quaternion at time t for the right foot
    acc_rot_right[t] = vqf_right.quatRotate(out_right["quat6D"][t], acc_right[t])

    # Rotate the gyroscope vector using the estimated quaternion at time t for the right foot
    gyr_rot_right[t] = vqf_right.quatRotate(out_right["quat6D"][t], gyr_right[t])
`````

- **Unit Conversion**: Convert units of the rotated acceleration and gyroscope data to be compatibale with `gaitmap` package requirements.


`````python
# Convert the rotated gyroscope data from radians per second back to degrees per second if needed
gyr_rot_left = gyr_rot_left * 180 / np.pi if mapping_type_units["GYRO"] == "deg/s" else gyr_rot_left
gyr_rot_right = gyr_rot_right * 180 / np.pi if mapping_type_units["GYRO"] == "deg/s" else gyr_rot_right

# Convert the rotated accelerometer data from meters per second squared back to 'g' if needed
acc_rot_left = acc_rot_left / 9.81 if mapping_type_units["ACCEL"] == "g" else acc_rot_left
acc_rot_right = acc_rot_right / 9.81 if mapping_type_units["ACCEL"] == "g" else acc_rot_right
`````

- **Combining Rotated Acceleration and Gyro Data**: 
   - Processes the rotated IMU data for both the left and right feet, organizing it into a structured format for further analysis.
   - The data is converted into a `pandas.DataFrame`, allowing urther processing with `gaitmap` package.


`````python
# Define the foot sensor names
left_foot = "left_foot"
right_foot = "right_foot"

# Create a dictionary for the rotated data for both left and right foot
rotated_data_dict = {
    "sensor": ["left_sensor"] * 6 + ["right_sensor"] * 6,
    "axis": [
        "acc_x", "acc_y", "acc_z", "gyr_x", "gyr_y", "gyr_z",  # Left foot axes
        "acc_x", "acc_y", "acc_z", "gyr_x", "gyr_y", "gyr_z"   # Right foot axes
    ]
}

# Create a pandas MultiIndex from the 'sensor' and 'axis'
multiindex_columns = pd.MultiIndex.from_arrays(
    [rotated_data_dict["sensor"], rotated_data_dict["axis"]],
    names=["sensor", "axis"]
)

# Combine the rotated IMU data into one array (left and right foot, side by side)
combined_data = np.hstack([acc_rot_left, gyr_rot_left, acc_rot_right, gyr_rot_right])

# Create the DataFrame using the MultiIndex for columns and the combined data for values
imu_data_rotated = pd.DataFrame(combined_data, columns=multiindex_columns)

# Display the first few rows of the rotated IMU data
print(imu_data_rotated.head())
`````

- **Convert to Foot-Frame**:
    - The rotated data is then converted to the foot-frame of reference, allowing us to analyze the movements of each foot in a consistent manner.

`````python
# Convert the rotated data to the foot-frame of reference
data_rotated = convert_to_fbf(imu_data_rotated, left_like="left_", right_like="right_")

# Displays the processed IMU data for both the left and right foot sensors
print(data_rotated)
`````

The table displays the processed IMU data for both the left and right foot sensors after applying the necessary transformations.

- **Acceleration Data**:
  - `acc_pa`: Acceleration in the anterior-posterior direction.
  - `acc_ml`: Acceleration in the medio-lateral direction.
  - `acc_si`: Acceleration in the superior-inferior direction.

- **Gyroscope Data**:
  - `gyr_pa`: Angular velocity in the anterior-posterior direction.
  - `gyr_ml`: Angular velocity in the medio-lateral direction.
  - `gyr_si`: Angular velocity in the superior-inferior direction.

**Applying the DTW**

In this section, we utilize Dynamic Time Warping (DTW) to analyze our IMU data and identify similarities between time-series sequences.

- **Initialization**: We start by initializing the Barth Dynamic Time Warping (DTW) algorithm. This step sets up the DTW framework for analyzing the data:

- **Segmenting the Data**: Once the DTW algorithm is initialized, we apply it to segment the foot-frame data using the IMU sampling rate. This process involves aligning the sequences and identifying temporal patterns in the data:

`````python
# Initialize the Barth Dynamic Time Warping (DTW) algorithm
dtw = BarthDtw()

# Apply DTW for segmenting the data using the IMU sampling rate
dtw = dtw.segment(data=data_rotated, sampling_rate_hz=fs)
`````

**Visualizing DTW Results**

To gain a deeper understanding of the results from our Dynamic Time Warping (DTW) analysis, we can plot additional information regarding the detected strides and the DTW cost function.

- Plot the gyroscope data of the left foot sensor, the DTW cost function, and highlight the identified strides.

`````python
# Define the sensor to be used for plotting (left foot sensor)
sensor = "left_sensor"

# Create subplots for gyroscope data and DTW cost function
fig, axs = plt.subplots(nrows=2, sharex=True, figsize=(12, 6))

# Plot the gyroscope data for the medio-lateral axis
dtw.data[sensor]["gyr_ml"].reset_index(drop=True).plot(ax=axs[0], color="blue")
axs[0].set_ylabel("Gyro [deg/s]")

# Plot the DTW cost function
axs[1].plot(dtw.cost_function_[sensor], color="purple")
axs[1].set_ylabel("DTW Cost [a.u.]")
axs[1].legend()

# Add stride annotations (green shaded areas) to the gyroscope and cost function plots
for _, s in dtw.stride_list_[sensor][["start", "end"]].iterrows():
    axs[0].axvspan(*s, alpha=0.3, color="green", label="Stride")  # Highlight strides in gyroscope plot
    axs[1].axvspan(*s, alpha=0.3, color="green")  # Highlight strides in cost function plot

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
`````

**Trajectory Reconstruction**

- The trajectory of each stride is reconstructed using double integration.
- The trajectory is estimated by calling the `estimate` method with the following parameters:
  - **data**: The pre-processed dataset (`dataset_sf`).
  - **stride_event_list**: The list of stride events (`ed.min_vel_event_list_`).
  - **sampling_rate_hz**: The sampling frequency of the data.


`````python
from gaitmap.trajectory_reconstruction import StrideLevelTrajectory

# Initialize the StrideLevelTrajectory object to handle trajectory reconstruction
trajectory = StrideLevelTrajectory()

# Estimate the trajectory of each stride using the double integration method
trajectory = trajectory.estimate(
    data=imu_data_rotated,                  # Rotated IMU data, which is used for trajectory reconstruction
    stride_event_list=ed.min_vel_event_list_,  # List of stride events containing minimal velocity information
    sampling_rate_hz=fs                     # Sampling frequency of the data
)

# The 'trajectory' object now contains the reconstructed trajectory for each stride based on the provided data and events.
`````

**Temporal Parameter Calculation**

Now that we have all the necessary information, we can calculate relevant temporal parameters, such as stride time, using the identified stride events.

- The primary output of this calculation is the temporal parameters, which provide insights into the timing of various phases of the gait cycle.
- This is achieved by initializing the TemporalParameterCalculation class, which includes methods to compute stride time and other temporal metrics.
- The calculation process involves using the min_vel_event_list_, which contains critical stride events, along with the sampling frequency to ensure accurate timing measurements.

`````python
from gaitmap.parameters import TemporalParameterCalculation

# Initialize the TemporalParameterCalculation object to handle temporal parameter calculations
temporal_paras = TemporalParameterCalculation()

# Calculate the temporal parameters using the identified stride events
temporal_paras = temporal_paras.calculate(
    stride_event_list=ed.min_vel_event_list_,  # List of stride events containing minimal velocity information
    sampling_rate_hz=fs                        # Sampling frequency of the data
)

# Access and print the pretty-printed temporal parameters for the left foot sensor
# This will provide a user-friendly display of the calculated parameters such as stride time, swing time, and stance time
temporal_paras.parameters_pretty_["left_sensor"]
`````

**Calculating Spatial Parameters**

We can also compute the spatial parameters relevant to the gait analysis:

- The spatial_paras object is initialized using the `SpatialParameterCalculation` class, which is specifically designed for calculating spatial parameters based on stride events.
- The calculate method is called with the following key inputs:
    - Stride Event List: This list (`ed.min_vel_event_list_`) provides the minimal velocity information necessary to define the start and end of each stride.
    - Positions: The `trajectory.position_` data contains the calculated positions of the foot during the gait cycle, allowing us to assess how far the foot travels with each stride.
    - Orientations: The `trajectory.orientation_` data contains the orientations of the foot, which can influence the overall gait mechanics.
    - Sampling Rate: The frequency of data collection (`sampling_rate_hz`) is also provided to ensure accurate calculations.
- After running this calculation, the `spatial_paras` object will contain important spatial metrics that can include:

`````python
from gaitmap.parameters import SpatialParameterCalculation

# Initialize the SpatialParameterCalculation object to handle spatial parameter calculations
spatial_paras = SpatialParameterCalculation()

# Calculate the spatial parameters using the identified stride events, positions, and orientations
spatial_paras = spatial_paras.calculate(
    stride_event_list=ed.min_vel_event_list_,  # List of stride events containing minimal velocity information
    positions=trajectory.position_,              # Trajectory positions calculated from the stride events
    orientations=trajectory.orientation_,        # Trajectory orientations calculated from the stride events
    sampling_rate_hz=fs           # Sampling frequency of the data
)

# The 'spatial_paras' object now contains the calculated spatial parameters for analysis
spatial_paras.parameters_pretty_["left_sensor"]
`````
